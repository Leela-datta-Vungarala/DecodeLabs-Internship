## Step 10 - Handle Missing Values

The missing values in the **CouponCode** column are replaced with **"NO_COUPON"** because customers may not have applied any coupon while placing their orders.

```python
df["CouponCode"] = df["CouponCode"].fillna("NO_COUPON")
```
